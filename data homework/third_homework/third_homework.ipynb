```python
class HashDict:
    def __init__(self, size: int):
        """
        size: 哈希表桶的数量（列表长度）
        """
        self.size = size
        # 哈希表：每个桶是列表，存储 (key, value) 元组
        self.table: list[list[tuple[str, int]]] = [[] for _ in range(size)]

    def hash_func(self, key: str) -> int:
        """
        对字符串key计算哈希值,再取模映射到桶下标
        return: 哈希表下标
        """
        # 内置哈希函数 + 取模保证落在 0 ~ size-1 范围内
        return abs(hash(key)) % self.size

    def put(self, key: str, value) -> None:
        """添加/更新键值对"""
        hash_result = self.hash_func(key)
        bucket = self.table[hash_result]

        # 遍历桶，找到相同key则更新值
        for i, (old_key, _) in enumerate(bucket):
            if key == old_key:
                bucket[i] = (key, value)
                return
        # 没找到则新增
        bucket.append((key, value))

    def find(self, key) -> int | None:
        """查找key对应的值,不存在返回None"""
        hash_result = self.hash_func(key)
        bucket = self.table[hash_result]

        for bucket_key, bucket_value in bucket:
            if key == bucket_key:
                return bucket_value
        return None

    def delete(self, key: str) -> None:
        """删除指定key"""
        hash_result = self.hash_func(key)
        bucket = self.table[hash_result]

        for i, (bucket_key, _) in enumerate(bucket):
            if key == bucket_key:
                del bucket[i]
                return
        return None

    def __str__(self):
        """打印哈希表内容"""
        items = {}
        for bucket in self.table:
            for key, value in bucket:
                items[key] = value
        return str(items)
```

```python
# 创建哈希表，10个桶
md = HashDict(10)

# 添加数据
md.put("apple", 100)
md.put("banana", 200)
md.put("cat", 300)

# 查找
print(md.find("apple"))   # 100
print(md.find("dog"))     # None

# 更新
md.put("apple", 999)

# 删除
md.delete("cat")

# 打印全部
print(md)  # {'apple': 999, 'banana': 200}
```
