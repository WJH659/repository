```python
import random

def quick_sort(arr):
    def _quick_sort(arr, low, high):
        if low < high:
            pi = partition(arr, low, high)
            _quick_sort(arr, low, pi - 1)
            _quick_sort(arr, pi + 1, high)
    _quick_sort(arr, 0, len(arr) - 1)
    return arr

def partition(arr, low, high):
    # 随机选择轴索引并交换到末尾
    rand_index = random.randint(low, high)
    arr[rand_index], arr[high] = arr[high], arr[rand_index]
    
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

# 测试用例
test_cases = [
    [3, 6, 8, 10, 1, 2, 1],
    [5, 3, 8, 4, 2, 7, 1, 10],
    [1, 2, 3, 4, 5],
    [10, 9, 8, 7, 6],
    [1],
    []
]

for arr in test_cases:
    print(f"原始数组: {arr}")
    sorted_arr = quick_sort(arr.copy())
    print(f"排序结果: {sorted_arr}\n")
```

```python
def quick_sort_3way(arr):
    def _quick_sort(arr, low, high):
        if low >= high:
            return
        pivot = arr[low]  # 选择第一个元素作为基准
        lt = low          # 小于区域边界（初始为low）
        gt = high         # 大于区域边界（初始为high）
        i = low + 1       # 当前遍历指针（从low+1开始）
        
        while i <= gt:
            if arr[i] < pivot:
                # 将小于基准的元素交换到左侧
                arr[lt], arr[i] = arr[i], arr[lt]
                lt += 1
                i += 1
            elif arr[i] > pivot:
                # 将大于基准的元素交换到右侧
                arr[i], arr[gt] = arr[gt], arr[i]
                gt -= 1
            else:
                # 等于基准的情况，直接移动指针
                i += 1
        
        # 递归排序左半部分（小于区域）和右半部分（大于区域）
        _quick_sort(arr, low, lt - 1)
        _quick_sort(arr, gt + 1, high)
    
    _quick_sort(arr, 0, len(arr) - 1)
    return arr

# 测试用例
test_cases = [
    [3, 6, 8, 10, 1, 2, 1, 1],  # 含重复元素
    [5, 3, 8, 4, 2, 7, 1, 10],
    [1, 2, 3, 4, 5],  # 已排序数组
    [10, 9, 8, 7, 6],  # 逆序数组
    [1],               # 单元素数组
    []                 # 空数组
]

for arr in test_cases:
    print(f"原始数组: {arr}")
    sorted_arr = quick_sort_3way(arr.copy())
    print(f"排序结果: {sorted_arr}\n")
```
