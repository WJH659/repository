```python
from collections.abc import Sequence
from typing import Any
class MySet:
    """
    创建一个类，能够实现集合的查询、插入、删除功能
    """
    def __init__(self,iterable: Sequence[Any]):
        self.iterable = list(iterable)
    def InsertSet(self,element:Any):
        """
        插入元素，element: 要插入的元素
        """
        for index,item in enumerate(self.iterable):
            if item == element:
                return False
        self.iterable.append(element)
        return True
    def contains(self,element:Any):
        """
        查询元素，element要查询的元素
        """
        for index,item in enumerate(self.iterable):
            if item == element:
                return True
        return False
    def DeleteSet(self,element:Any):
        """
        删除元素，element要删除的元素
        """
        if element not in self.iterable:
            return False
        self.iterable.remove(element)
        return True
```

```python
set1={1,2,3,4,5}
set1=MySet(set1)
print(set1.DeleteSet(5))
print(set1.contains(5))
print(set1.contains(4))
print(set1.InsertSet(99))
print(set1.iterable)
```
