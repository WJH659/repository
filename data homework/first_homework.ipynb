```python
from typing import Iterable,Any
class DeleteItem:
    """在列表中删除元素
    """
    def __init__(self,iterable: Iterable[Any]):
        """param iterable: 可迭代对象，在其内部删除元素"""
        self.iterable = list(iterable)
    def delete(self,index_to_delete: int) -> list[Any]:
        """删除元素
        :param index_to_delete : 要删除的索引"""
        length: int = len(self.iterable)
        if index_to_delete >=length or index_to_delete <0:
            raise IndexError("输入的数字超出范围")
        for i in range(index_to_delete,length-1,1):
            self.iterable[i]=self.iterable[i+1]
        self.iterable.pop()
        return self.iterable
```

```python
f=DeleteItem([0,1,2,3,4,5])
print(f.delete(2))  #删除中间元素
se = DeleteItem([18,26,35,44])
print(se.delete(0)) #删除第一个元素
```
