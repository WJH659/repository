```python
class ListNode:
    """链表节点类"""
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class LinkedList:
    """单链表类，包含反转、检测环功能"""
    def __init__(self):
        self.head = None

    def add_at_tail(self, val):
        """向链表尾部添加节点（辅助方法，用于测试）"""
        new_node = ListNode(val)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def print_list(self):
        """打印链表（辅助方法，用于测试）"""
        current = self.head
        vals = []
        while current:
            vals.append(str(current.val))
            current = current.next
        print(" -> ".join(vals))

    def reverse(self):
        """作业1：迭代法反转链表（时间复杂度O(n)，空间复杂度O(1)）"""
        prev = None
        curr = self.head
        while curr:
            # 保存下一个节点
            next_node = curr.next
            # 反转当前节点的指针
            curr.next = prev
            # 移动指针
            prev = curr
            curr = next_node
        # 更新链表头
        self.head = prev

    def has_cycle(self):
        """作业2：快慢指针法检测环（Floyd判圈算法）"""
        slow = self.head
        fast = self.head
        while fast and fast.next:
            slow = slow.next          # 慢指针走1步
            fast = fast.next.next     # 快指针走2步
            if slow == fast:          # 相遇说明有环
                return True
        return False
```

```python
# ------------------- 测试用例 -------------------
if __name__ == "__main__":
    print("===== 测试1：链表反转 =====")
    # 创建链表 [1, 2, 3, 4, 5]
    ll1 = LinkedList()
    for num in [1, 2, 3, 4, 5]:
        ll1.add_at_tail(num)
    print("反转前：", end="")
    ll1.print_list()
    # 调用反转方法
    ll1.reverse()
    print("反转后：", end="")
    ll1.print_list()

    print("\n===== 测试2：无环链表检测 =====")
    ll2 = LinkedList()
    for num in [10, 20, 30]:
        ll2.add_at_tail(num)
    print("链表 [10,20,30] 是否有环：", ll2.has_cycle())  # 输出 False

    print("\n===== 测试3：有环链表检测 =====")
    ll3 = LinkedList()
    node1 = ListNode(1)
    node2 = ListNode(2)
    node3 = ListNode(3)
    node1.next = node2
    node2.next = node3
    node3.next = node1  # 尾节点指向头节点，形成环
    ll3.head = node1
    print("循环链表是否有环：", ll3.has_cycle())  # 输出 True
```
